```python
ruta = "peru.jpg"
```

```python
imagen = cv2.imread(ruta)
imagen_rgb = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
imagen_gray = cv2.cvtColor(imagen_rgb,cv2.COLOR_RGB2GRAY)
imagen_gray2 = cv2.cvtColor(imagen_gray,cv2.COLOR_GRAY2RGB)
```

```python
plt.imshow(imagen_gray2)
plt.show()
```

```python
R = imagen_rgb[:,:,0].astype(int)
G = imagen_rgb[:,:,1].astype(int)
B = imagen_rgb[:,:,2].astype(int)
```

```python
type(R[0,0])
```

```python
EG = (R+G+B)/3
```
