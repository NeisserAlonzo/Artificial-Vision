# OPERADORES PUNTUALES

```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Identidad

```python
img = cv2.imread("identidad.png")
```

```python
img_identidad = img
```

```python
f, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)
ax1.imshow(img)
ax1.set_title('Original')

ax2.imshow(img_identidad)
ax2.set_title('Identidad')

plt.show()
```

## Negativo

```python
img1 = cv2.imread("negativo.png")
```

```python
img_negativo = 255-img1
```

```python
f, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)
ax1.imshow(img1)
ax1.set_title('Original')

ax2.imshow(img_negativo)
ax2.set_title('Negativo')

plt.show()
```

## Umbral 

```python
img2 = cv2.imread("umbral.png")
```

```python
img_umbral = np.zeros(img2.shape)
```

```python
umbral = 150
```

```python
img_umbral[img2 > umbral] = 255
img_umbral[img2 <= umbral] = 0
```

```python
f, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)
ax1.imshow(img2)
ax1.set_title('Original')

ax2.imshow(img_umbral)
ax2.set_title('Umbral '+str(umbral))

plt.show()
```

## Umbral Binario

```python
img3 = cv2.imread("umbral_binario.png")
```

```python
img_umbral_binario = np.zeros(img3.shape)
```

```python
umbral_1 = 40
umbral_2 = 190
```

```python
img_umbral_binario[img3 <= umbral_1] = 255
img_umbral_binario[img3 >= umbral_2] = 255
img_umbral_binario[img3 > umbral_1] = 0
img_umbral_binario[img3 < umbral_2] = 0
```

```python
f, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)
ax1.imshow(img3)
ax1.set_title('Original')

ax2.imshow(img_umbral_binario)
ax2.set_title('Umbral '+str(umbral_1) + " y " +str(umbral_2))

plt.show()
```

## Umbral Binario Invertido

```python
img4 = cv2.imread("binario_invertido.png")
```

```python
img_umbral_binario_invertido = np.zeros(img4.shape)
```

```python
umbral_1 = 40
umbral_2 = 190
```

```python
img_umbral_binario_invertido[img4 <= umbral_1] = 0
img_umbral_binario_invertido[img4 >= umbral_2] = 0
img_umbral_binario_invertido[img4 > umbral_1] = 255
img_umbral_binario_invertido[img4 < umbral_2] = 255
```

```python
f, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)
ax1.imshow(img3)
ax1.set_title('Original')

ax2.imshow(img_umbral_binario_invertido)
ax2.set_title('Umbral '+str(umbral_1) + " y " +str(umbral_2))

plt.show()
```

## Extensin 

```python
img6 = cv2.imread("extension.png")
```

```python
img_extension = np.zeros(img6.shape)
```

```python
umbral_1 = 40
umbral_2 = 190
```

```python
img_extension[img6 <= umbral_1] = 255
img_extension[img6 >= umbral_2] = 255
img_extension[img6 > umbral_1] = ((img6-umbral_1)*255)/(umbral_2-umbral_1)
img_extension[img6 < umbral_2] = ((img6-umbral_1)*255)/(umbral_2-umbral_1)
```
